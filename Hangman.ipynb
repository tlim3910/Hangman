```python
%matplotlib notebook
from numpy import random, append
```

```python
WORDLIST_FILENAME = "words.txt"
```

```python
def load_words():
    """
    Returns a list of valid words. Words are strings of lowercase letters.
    
    Depending on the size of the word list, this function may
    take a while to finish.
    """
    print('')
    print("Loading word list from file...")
    inFile = open(WORDLIST_FILENAME, 'r')
    line = inFile.readline()
    wordlist = line.split()
    print("  ", len(wordlist), "words loaded.")
    
    return wordlist
```

```python
def choose_word(wordlist):
    """
    wordlist (list): list of words (strings)
    
    Returns a word from wordlist at random
    """
    return random.choice(wordlist)
```

```python
def letters_guessed(letter_guessed, secret_word, where):
    length = len(secret_word)
    i = 0
    while i in range(length):
        if letter_guessed == secret_word[i]:
            where[str(2*i)] = letter_guessed
        i += 1
    
    return where
```

```python
def update(where, secret_word):
    length = 2*len(secret_word) - 1
    guessed_word = ''
    for i in range(length):
        if i%2 == 0:
            j = str(i)
            if j in where:
                add = guessed_word
                guessed_word = add + where[j]
            else:
                add = guessed_word
                guessed_word = add + '_'
        if i%2 == 1:
            add = guessed_word
            guessed_word = add + ' '
    
    print(guessed_word)
    
    return guessed_word
```

```python
def get_available_letters(letter_guessed, letters_available):
    '''
    letters_guessed: list (of letters), which letters have been guessed so far
    returns: string (of letters), comprised of letters that represents which letters have not
      yet been guessed.
    '''
    available_v2 = letters_available
    letters_available = ''
    for i in range(len(available_v2)):
        if letter_guessed != available_v2[i]:
            add = letters_available
            letters_available = add + available_v2[i]
        
    return letters_available
```

```python
def word_guessed(guessed_word):
    flag = True
    for i in range(len(guessed_word)):
        if guessed_word[i] == '_':
            flag = False

            break
    
    return flag
```

```python
def hangman():
    wordlist = load_words()
    secret_word = choose_word(wordlist)
    letters_available = 'abcdefghijklmnopqrstuvwxyz'
    choices_left = 20
    flag = False
    where = {}
    while choices_left > 0 and flag == False:
        print('')
        print('You have ', choices_left, ' guesses left.') 
        print('Available letters: ', letters_available)
        letter_guessed = input('Please guess a letter: ')
        print('')
        where = letters_guessed(letter_guessed, secret_word, where)
        guessed_word = update(where, secret_word)
        letters_available = get_available_letters(letter_guessed, letters_available)
        flag = word_guessed(guessed_word)
        choices_left -= 1
        print('')
        print('----------------------------------------------')
    
    print('Your secret word was', secret_word, '.')
    
    return       
```
